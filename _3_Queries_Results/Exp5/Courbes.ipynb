```python
import pandas as pd
import matplotlib.pyplot as plt

# Charger les données
file_path = "CozoRPG_ABS.xlsx"
df = pd.read_excel(file_path)

# S'assurer que la colonne '#abstraction_level' est bien entière
df['#abstraction_level'] = df['#abstraction_level'].astype(int)

# Créer la figure
plt.figure(figsize=(14, 6))

# Tracer uniquement la courbe CozoRPG
plt.plot(df['#abstraction_level'], df['CozoRPG'], marker='o', color='green', label='CozoRPG')

# Ajouter les titres et les axes
plt.title("Évolution de CozoRPG", fontsize=16)
plt.xlabel("Niveau d'abstraction (#abstraction_level)", fontsize=12)
plt.ylabel("Temps d'exécution (ms)", fontsize=12)
plt.xticks(df['#abstraction_level'])  # Forcer les abscisses à être entiers
plt.legend()
plt.grid(True)
plt.tight_layout()

# Sauvegarder le graphique
plt.savefig("CozoRPG_ABS.png")
plt.show()
```
