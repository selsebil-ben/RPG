```python
import pandas as pd
import matplotlib.pyplot as plt


file_path = "CozoPG_vs_CozoRPG_R1.xlsx"

df = pd.read_excel(file_path)


plt.figure(figsize=(14, 6))
plt.plot(df['DS_size'], df['CozoPG'], marker='o', color='blue', label='CozoPG')
plt.plot(df['DS_size'], df['CozoRPG'], marker='s', color='pink', label='CozoRPG')

# Ajouter titre et axes
plt.title("Temps d'exécution : CozoPG vs CozoRPG", fontsize=16)
plt.xlabel("Taille des données (DS_size)", fontsize=12)
plt.ylabel("Temps d'exécution (ms)", fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.legend()
plt.grid(True)
plt.tight_layout()

# Sauvegarder le graphique
plt.savefig("cozoPG_vs_cozoRPG_R1.png")
plt.show()
```
