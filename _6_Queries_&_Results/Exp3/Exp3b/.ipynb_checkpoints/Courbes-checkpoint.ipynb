```python

#Q1 Retrieve nodes with their labels and properties belonging to a given reified node.
?[associated_node, associated_label, associated_prop] :=
*n_composed_of_node_label[id_rn, associated_node, associated_label],
*n_composed_of_node_prop[id_rn, associated_node, associated_prop],
id_rn = 400016
```

```python
#Visualization Q1 

import pandas as pd
import matplotlib.pyplot as plt

# Charger les donnes
file_path = "CozoRPG_RN_Q1.xlsx"
df = pd.read_excel(file_path)

# Convertir la colonne '#reified edges' en chane de caractres
df['#reified nodes'] = df['#reified nodes'].astype(str)

# Crer la figure
plt.figure(figsize=(14, 6))

# Tracer uniquement la courbe CozoRPG
plt.plot(df['#reified nodes'], df['CozoRPG'], marker='o', color='green', label='RPG (CozoDB)')

# Ajouter les titres et les axes
# plt.title("Execution Time in CozoDB (RPG)", fontsize=16)
plt.xlabel("#reified nodes", fontsize=17)
plt.ylabel("Execution Time (ms)", fontsize=17)
plt.xticks(rotation=30, ha='right')
plt.legend(fontsize=16, loc="best")
plt.grid(True)
plt.tight_layout()
# Taille plus grande pour les ticks de laxe des X
plt.xticks(rotation=45, ha='right', fontsize=19)

# agrandir ceux de laxe Y
plt.yticks(fontsize=19)
# Sauvegarder le graphique
plt.savefig("CozoRPG_RN_Q1.png", dpi=300, bbox_inches="tight")
plt.show()

```
