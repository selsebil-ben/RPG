```python
#Visualization Q1 Q3

import pandas as pd
import matplotlib.pyplot as plt

file_path = "CozoRPG_Q1_Q3.xlsx"
df = pd.read_excel(file_path)

plt.figure(figsize=(14, 6))

# Tracer les courbes
plt.plot(df['DS_size'], df['Q1'], marker='o', color='green', label='Q1')
plt.plot(df['DS_size'], df['Q3'], marker='^', color='red', label='Q3')

# Choisir les labels que tu veux garder visibles
labels_to_show = ['100K_500K',
 '500K_2M500K',
 '900K_4M500K',
 '1M300K_6M500K',
 '1M700K_8M500K',
 '2M100K_10M500K',
 '2M500K_12M500K',
 '2M900K_14M500K']


# Construire la liste des ticks
xticks = range(len(df['DS_size']))
xticklabels = [lbl if lbl in labels_to_show else "" for lbl in df['DS_size']]

plt.xticks(ticks=xticks, labels=xticklabels, rotation=30, ha='right', fontsize=19)
plt.yticks(fontsize=19)
# Ajouter titre et axes
# plt.title("Execution Time: Q1 vs Q2", fontsize=16)
plt.xlabel("Dataset size", fontsize=17)
plt.ylabel("Execution Time (ms)", fontsize=17)

# Lgende avec texte plus grand
plt.legend(fontsize=16, loc="best")
plt.grid(True)
plt.tight_layout()

# Sauvegarder le graphique
plt.savefig("exp3aQ1Q3.png", dpi=300)
plt.show()
```
