```python
#Visualization
import pandas as pd
import matplotlib.pyplot as plt

# Charger les donnes
file_path = "CozoRPG_ABS.xlsx"
df = pd.read_excel(file_path)

# S'assurer que la colonne '#abstraction_level' est bien entire
df['#abstraction_level'] = df['#abstraction_level'].astype(int)

# Crer la figure
plt.figure(figsize=(14, 6))

# Tracer uniquement la courbe CozoRPG
plt.plot(df['#abstraction_level'], df['CozoRPG'], marker='o', color='green', label='RPG (CozoDB)')

# Ajouter les titres et les axes
# plt.title("Execution Time in CozoDB (RPG)", fontsize=16)
plt.xlabel("#abstraction level", fontsize=17)
plt.ylabel("Execution Time (ms)", fontsize=17)
plt.xticks(df['#abstraction_level'])  # Forcer les abscisses  tre entiers
plt.legend(fontsize=16, loc="best")
plt.grid(True)
plt.tight_layout()
# Taille plus grande pour les ticks de laxe des X
plt.xticks(ha='right', fontsize=20)

# agrandir ceux de laxe Y
plt.yticks(fontsize=20)
# Sauvegarder le graphique
plt.savefig("CozoRPG_ABS.png", dpi=300, bbox_inches="tight")
plt.show()
```
