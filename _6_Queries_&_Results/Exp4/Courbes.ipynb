```python

#Q1 Retrieve nodes with their labels and properties belonging to a given reified node.
#cozo 
?[associated_node, associated_label, associated_prop] :=
*n_composed_of_node_label[id_rn, associated_node, associated_label],
*n_composed_of_node_prop[id_rn, associated_node, associated_prop],
id_rn = 400016

#Neo4j
MATCH (tc:TAGGEDCOMMENT {id: 400056})-[:_node_membership]->(tag:TAG)
MATCH (tc:TAGGEDCOMMENT {id: 400056})-[:_node_membership]->(n)
RETURN tag.name AS tag_name, labels(n) AS node_types
```

```python
#visualisation Q1
import pandas as pd
import matplotlib.pyplot as plt


file_path = "Cozo_vs_HOGDB _Q1.xlsx"

df = pd.read_excel(file_path)


plt.figure(figsize=(14, 6))
plt.plot(df['#reified nodes'], df['Cozo'], marker='o', color='green', label='RPG (CozoDB)')
plt.plot(df['#reified nodes'], df['HOGDB'], marker='^', color='red', label='HO-GDB (Neo4j)')

# Ajouter titre et axes
# plt.title("Execution Time : Cozo vs HOGDB", fontsize=16)
plt.xlabel("# reified nodes (# subgraphs)", fontsize=17)
plt.ylabel("Execution Time (ms)", fontsize=17)
plt.xticks(rotation=30, ha='right')
plt.legend(fontsize=16, loc="best")
plt.grid(True)
plt.tight_layout()
# Taille plus grande pour les ticks de laxe des X
plt.xticks(ha='right', fontsize=19)

# agrandir ceux de laxe Y
plt.yticks(fontsize=19)
# Sauvegarder le graphique
plt.savefig("graph_cozo_vs_HOGDB_Q1.png", dpi=300, bbox_inches="tight")
plt.show()
```
